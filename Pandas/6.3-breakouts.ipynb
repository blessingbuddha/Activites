1. Use the given x and y axis values to create a line graph
    - other instuctions are written in as comments 

```python
#import Packages 
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 

# x and y axis values
x = [1,2,3,4,5,6]
y = [2,4,1,5,2,6]

# plotting the points

plt.plot(
    x,
    y, 
    # label = '2x', 
    color = 'green', # change the line color to green
    linewidth = 3, # change linewidth to 3
    linestyle = '-', # change line style to dashed 
    marker = 'o', # change the marker style (use documentation and pick one)
    markersize=12, # change marker size to 12
    markerfacecolor= 'blue' # change marker color to blue 
    )

#tweak  

    # setting x and y axis range to 0,8 using x/y lim function

plt.xlim(0,8)
plt.ylim(0,8)


    # naming the x  and y axis 

plt.xlabel('X Axis',fontdict={'fontname':'Comic Sans MS'} )
plt.ylabel('Y Axis',fontdict={'fontname':'Comic Sans MS'})

# add a  a title to my graph

plt.title('Breakout 1 -- Question 1', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})

# render plot 
plt.show()
```

2.Use 'gas_prices.csv' and create a line graph that shows the gas prices over time for all of the European countries (France,Germany,Italy,UK).

```python
# import packages

import pandas as pd
import matplotlib.pyplot as plt 

# prepare data

gas = pd.read_csv('../dap-curriculum/Section06/resources/gas_prices.csv')




# plot the four countries 

plt.plot(gas.Year, gas.USA, 'r.-', label = 'France')
plt.plot(gas.Year,gas.Canada, 'b.-',label = 'Germany')
plt.plot(gas.Year,gas.UK, 'g.-', label = 'Italy')
plt.plot(gas.Year,gas.Australia, 'y.-', label = 'Austalia')



# change the x ticks

plt.xticks(gas.Year[::3])

# add X and Y labels 

plt.ylabel('Price per Gallon (USD)')
plt.xlabel('Country')

# add a title to the Plot 

plt.title('Gas Prices Over Time', fontdict={'fontweight': 'bold', 'fontsize':14})

# add a legend 

plt.legend()

# save your graph (if you want to)
plt.savefig('GasPlot.png', dpi=300)

# render plot 

```
